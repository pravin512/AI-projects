```python
#Why?
#Sometimes data doesn't fit into straight line. We can use polynomial regression to fit a curve to the data.

#Features to be considered: x^2, x^3, x^4, x^5....... to capture the curves(non-linear) in the data
```

```python
#lets generate curve(Non-linear) data
import numpy as np
import matplotlib.pyplot as plt

# Generate non-linear data
np.random.seed(42)
x = np.linspace(0, 10, 100).reshape(-1, 1)  # 100 points between 0 and 10
y = 2 + 3*x - 0.5*x**2 + np.random.randn(100, 1) * 2  # Quadratic equation + noise

# Plot the dataset
plt.scatter(x, y)
# plt.xlabel("X")
# plt.ylabel("y")
# plt.title("Non-Linear Data")
# plt.legend()
plt.show()
```

```python
#fit a simple linear regression model(Baseline)
#lets see how linear regression performs on curve(non-linear) data
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
y_pred_linear = lin_reg.predict(x)

# Plot the results
plt.scatter(x, y, label="Data")
plt.plot(x, y_pred_linear, color="red", label="Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression on Non-Linear Data")
plt.legend()
plt.show()
```

```python
# Apply Polynomial Regression(Degree=2)
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) #degree=2 means x^2(quadratic)
x_poly = poly_reg.fit_transform(x)
poly_reg.fit(x_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
y_pred_poly = lin_reg_2.predict(x_poly)

# Plot the results
plt.scatter(x, y, label="Data")
plt.plot(x, y_pred_poly, color="green", label="Polynomial Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression on Non-Linear Data")
plt.legend()
plt.show()
```

```python
# lets try higher degrees(More complex models)
#les experiment with different degrees(3,4,5) & see how they perform
degrees = [2, 3, 5]

plt.scatter(x, y, color="blue", label="Actual Data")
for degree in degrees:
    poly_reg = PolynomialFeatures(degree=degree)
    x_poly = poly_reg.fit_transform(x)
    lin_reg = LinearRegression()
    lin_reg.fit(x_poly, y)
    y_pred_poly = lin_reg.predict(x_poly)
    plt.plot(x, y_pred_poly, label=f"Polynomial Regression (Degree={degree})")
    

plt.xlabel("X")
plt.ylabel("y")
plt.title("Effect of different degrees on Polynomial Regression")
plt.legend()
plt.show()
```
