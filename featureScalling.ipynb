```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Generate sample data
np.random.seed(42)
X = np.random.rand(100, 2) * [100, 500]  # First column (0-100), second column (0-500)

# Standardization
scaler_standard = StandardScaler()
X_standardized = scaler_standard.fit_transform(X)

# Min-Max Normalization
scaler_minmax = MinMaxScaler()
X_normalized = scaler_minmax.fit_transform(X)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot original data
axes[0].scatter(X[:, 0], X[:, 1], color='blue')
axes[0].set_title("Original Data")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# Plot Standardized Data
axes[1].scatter(X_standardized[:, 0], X_standardized[:, 1], color='red')
axes[1].set_title("Standardized Data (Mean=0, Std=1)")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")

# Plot Min-Max Normalized Data
axes[2].scatter(X_normalized[:, 0], X_normalized[:, 1], color='green')
axes[2].set_title("Min-Max Scaled Data (0-1 Range)")
axes[2].set_xlabel("Feature 1")
axes[2].set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```
